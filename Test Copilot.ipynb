```python


import pygame
import random
import numpy as np


from IPython import get_ipython #pour l'affichage des graphes dans une fenêtre hors IDE

get_ipython().run_line_magic('matplotlib', 'qt') #affichage des graphes dans une fenêtres hors IDE



# Initialize Pygame
pygame.init()

# Set up display
width, height = 800, 600
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("Réaction entre l'acide chlorhydrique et l'hydroxyde de sodium")
#print("alive")
# Define molecule class
class Molecule:
    def __init__(self,identity=1):
        if identity == 1 :
            self.radius = random.randint(5, 20)
            self.color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
            self.x = random.randint(self.radius, width - self.radius)
            self.y = random.randint(self.radius, height - self.radius)
            self.vx = random.uniform(-1, 1)/20
            self.vy = random.uniform(-1, 1)/20
            self.mass = 1
            
        else :
            self.radius = 10
            self.color = (255, 0, 0)
            self.x = random.randint(self.radius, width - self.radius)
            self.y = random.randint(self.radius, height - self.radius)
            self.vx = random.uniform(-1, 1)/20
            self.vy = random.uniform(-1, 1)/20
            self.mass = 5
        
    def addPosition(self, posx,posy):
        self.x += posx
        self.y += posy
        
    def addVelocity(self, velx,vely):
        self.vx += velx
        self.vy += vely

    def move(self):
        self.x += self.vx
        self.y += self.vy

        # Bounce off walls
        if self.x <= self.radius or self.x >= width - self.radius:
            self.vx *= -1
        if self.y <= self.radius or self.y >= height - self.radius:
            self.vy *= -1
            
    def collide(self,p1,p2):
        dXx = p1.x-p2.x
        dXy = p1.y-p2.y
        dist = np.sqrt(dXx**2 + dXy**2)
        if dist < p1.radius+p2.radius:
            offset = dist-(p1.radius+p2.radius)
            p1.addPosition((-dXx/dist)*offset/2,(-dXy/dist)*offset/2)
            p2.addPosition((dXx/dist)*offset/2,(dXy/dist)*offset/2)
            total_mass = p1.mass+p2.mass
            #dv1 = -2*p2.mass/total_mass*np.inner(p1.V-p2.V,p1.X-p2.X)/np.sum((p1.X-p2.X)**2)*(p1.X-p2.X)
            scalprod1 = (p1.vx-p2.vx)*(p1.x-p2.x)+ (p1.vy-p2.vy)*(p1.y-p2.y)
            scalprod2 = (p2.vx-p1.vx)*(p2.x-p1.x)+ (p2.vy-p1.vy)*(p2.y-p1.y)
            
            dv1x = -2*p2.mass/total_mass*scalprod1/((p1.x-p2.x)**2+(p1.y-p2.y)**2)*(p1.x-p2.x)
            dv1y = -2*p2.mass/total_mass*scalprod1/((p1.x-p2.x)**2+(p1.y-p2.y)**2)*(p1.y-p2.y)
            
            dv2x = -2*p1.mass/total_mass*scalprod2/((p2.x-p1.x)**2+(p2.y-p1.y)**2)*(p2.x-p1.x)
            dv2y = -2*p1.mass/total_mass*scalprod2/((p2.x-p1.x)**2+(p2.y-p1.y)**2)*(p2.y-p1.y)
            
            
            p1.addVelocity(dv1x,dv1y)
            p2.addVelocity(dv2x,dv2y)
    
        

# Create molecules
num_molecules = 30
molecules = [Molecule() for _ in range(10)] + [Molecule(0) for _ in range(num_molecules-10)]
#molecules = [Molecule() for _ in range(10)]+[Molecule(0) for _ in range(10)]

# Main loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Update molecule positions
    for molecule in molecules:
        molecule.move()
        for molecule2 in molecules :
            if molecule2!=molecule :
                molecule.collide(molecule,molecule2)

    # Draw molecules
    screen.fill((255, 255, 255))
    for molecule in molecules:
        pygame.draw.circle(screen, (0,0,0), (int(molecule.x), int(molecule.y)), molecule.radius+1)
        pygame.draw.circle(screen, molecule.color, (int(molecule.x), int(molecule.y)), molecule.radius-1)

    pygame.display.flip()

# Clean up
pygame.quit()
```

```python
import pygame
import random

# Initialize Pygame
pygame.init()

# Constants
WIDTH, HEIGHT = 800, 600
MOLECULE_RADIUS = 10
NUM_MOLECULES = 50
MOLECULE_COLORS = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]  # Red, Green, Blue

# Create a window
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Colliding Molecules Simulation")

# Create random molecules
molecules = []
for _ in range(NUM_MOLECULES):
    x = random.randint(MOLECULE_RADIUS, WIDTH - MOLECULE_RADIUS)
    y = random.randint(MOLECULE_RADIUS, HEIGHT - MOLECULE_RADIUS)
    dx = random.uniform(-1, 1)
    dy = random.uniform(-1, 1)
    color = random.choice(MOLECULE_COLORS)
    molecules.append((x, y, dx, dy, color))

# Main loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Update molecule positions
    for i, (x, y, dx, dy, _) in enumerate(molecules):
        x += dx
        y += dy
        molecules[i] = (x, y, dx, dy, MOLECULE_COLORS[i % len(MOLECULE_COLORS)])

    # Detect collisions (simple bounding box check)
    for i, (x1, y1, _, _, _) in enumerate(molecules):
        for j, (x2, y2, _, _, _) in enumerate(molecules):
            if i != j:
                distance = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
                if distance < 2 * MOLECULE_RADIUS:
                    # Swap velocities (elastic collision)
                    molecules[i] = (x1, y1, -dx, -dy, MOLECULE_COLORS[i % len(MOLECULE_COLORS)])
                    molecules[j] = (x2, y2, -dx, -dy, MOLECULE_COLORS[j % len(MOLECULE_COLORS)])

    # Draw molecules
    screen.fill((255, 255, 255))  # Clear screen
    for x, y, _, _, color in molecules:
        pygame.draw.circle(screen, color, (int(x), int(y)), MOLECULE_RADIUS)

    pygame.display.flip()

# Clean up
pygame.quit()
```

```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Set up the window
win = pygame.display.set_mode((500, 500))
pygame.display.set_caption("Pygame Example")

# Define colors
black = (0, 0, 0)
white = (255, 255, 255)

# Create a font
sysfont = pygame.font.SysFont(None, 50)

# Main game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Clear the screen
    win.fill(white)

    # Render the text
    text_surface = sysfont.render("Hello, Pygame!", True, black)

    # Center the text on the screen
    text_rect = text_surface.get_rect(center=(250, 250))

    # Blit the text onto the window
    win.blit(text_surface, text_rect)

    # Update the display
    pygame.display.update()

# Clean up and quit
pygame.quit()
```
